```python
import requests
import logging
import time
from tqdm.auto import tqdm
from urllib.parse import quote
from collections import defaultdict
import ast
import pandas as pd

class LitVarFetcher:
    def __init__(self, gene_name: str):
        self.gene_name = gene_name
        self.session = requests.Session()
        logging.basicConfig(level=logging.INFO)

    # 1) Gene-based variant list
    def fetch_gene_variants(self):
        base_url = "https://www.ncbi.nlm.nih.gov/research/litvar2-api/variant/search/gene/"
        url = base_url + self.gene_name

        try:
            response = self.session.get(url, timeout=20)
            response.raise_for_status()
        except Exception as e:
            logging.error(f"LitVar gene search API 오류: {e}")
            return []

        variants = []
        lines = response.text.strip().split("\n")

        for line in lines:
            try:
                d = ast.literal_eval(line.strip())
                variants.append(d)
            except Exception:
                continue

        return variants

    # 2) PMID list for each variant
    def fetch_pmids_for_variant(self, variant_id: str):
        base_url = "https://www.ncbi.nlm.nih.gov/research/litvar2-api/variant/get/"
        encoded = quote(variant_id)
        # url = f"{base_url}{encoded}?format=api"
        url= f"{base_url}{encoded}/publications"

        try:
            r = self.session.get(url, timeout=10)
            r.raise_for_status()
            j = r.json()
            pmids = j.get("pmids", [])
            if not pmids:
                logging.info(f"{variant_id}에는 PMID가 없습니다.")
            return pmids
        except Exception as e:
            logging.warning(f"PMID 조회 실패 ({encoded}): {e}")
            return []


    # 2.2) hgvsc list for each variant
    def fetch_hgvs_for_variant(self, variant_id: str):
        base_url = "https://www.ncbi.nlm.nih.gov/research/litvar2-api/variant/get/"
        encoded = quote(variant_id)
        # url = f"{base_url}{encoded}?format=api"
        url= f"{base_url}{encoded}"

        try:
            r = self.session.get(url, timeout=10)
            r.raise_for_status()
            j = r.json()
            hgvs = j.get("hgvs", [])
            if not hgvs:
                logging.info(f"{variant_id}에는 hgvs가 없습니다.")
            return hgvs
        except Exception as e:
            logging.warning(f"hgvs 조회 실패 ({encoded}): {e}")
            return []
    # 3) Build map and DataFrame
    def build_variant_dataframe(self):
        variants = self.fetch_gene_variants()
        print(variants)
        if not variants:
            logging.error("변이 정보를 가져오지 못했습니다.")
            return pd.DataFrame()

        rows = []
        pmid_to_variants = defaultdict(list)


        for v in tqdm(variants, desc="PMID 매핑 중"):
            id = v.get("_id")
            rsid = v.get("rsid")
            gene_list = v.get("gene", [])


            pmids = self.fetch_pmids_for_variant(id)
            hgvs = self.fetch_hgvs_for_variant(id)

            if not id:
                continue

            variant_label = rsid if rsid else id
            pmids = self.fetch_pmids_for_variant(id)
            hgvs = self.fetch_hgvs_for_variant(id)

            if not pmids:
                rows.append({
                    "_id": id,
                    "rsid": rsid,
                    "gene": ",".join(gene_list),
                    "pmid": None,
                    "hgvs": hgvs if hgvs else None
                })
            else:
                for pmid in pmids:
                    rows.append({
                        "_id": id,
                        "rsid": rsid,
                        "gene": ",".join(gene_list),
                        "pmid": int(pmid),
                        "hgvs": hgvs if hgvs else None
                    })


            time.sleep(0.3)  # API 부담 완화

        df = pd.DataFrame(rows)
        return df

# 실행
gene = "EEF1A2"
fetcher = LitVarFetcher(gene)
litvar_df = fetcher.build_variant_dataframe()
```

```python
#litvar hgvsc, hgvsp나누기
## hgvsc 확인

import pandas as pd
import requests
from tqdm import tqdm

# -------------------------
# Helper functions
# -------------------------

def get_dbsnp_hgvs_list(rsid):
    """
    dbSNP RefSNP API에서 rsID에 해당하는 모든 c.HGVS 리스트 가져오기
    """
    rs_num = rsid.replace("rs", "")
    url = f"https://api.ncbi.nlm.nih.gov/variation/v0/beta/refsnp/{rs_num}"
    r = requests.get(url, timeout=10)
    if r.status_code != 200:
        return []

    data = r.json()
    hgvs_list = set()
    primary_snapshot_data = data.get("primary_snapshot_data", {})

    # alleles에 대한 HGVS 정보 추출
    if "placements_with_allele" in primary_snapshot_data:
        for placement in primary_snapshot_data["placements_with_allele"]:
            for allele in placement.get("alleles", []):
                for hgvs in allele.get("hgvs", []):
                    if "coding" in hgvs:
                        hgvs_list.add(hgvs["coding"])
    return list(hgvs_list)


def check_hgvs_in_dbsnp(rsid, hgvs):
    """
    LitVar HGVS가 dbSNP HGVS 리스트에 존재하는지 확인
    """
    hgvs_list = get_dbsnp_hgvs_list(rsid)
    exists = hgvs in hgvs_list
    return exists, hgvs_list


# -------------------------
# Load LitVar file
# -------------------------
df = pd.read_csv("/content/sample_data/pmid_hgvs.txt", sep="\t", dtype=str).fillna("")

results = []

for _, row in tqdm(df.iterrows(), total=len(df)):
    rsid = row.get("rsid", "").strip()
    hgvs = row.get("hgvs", "").strip()

    if not rsid or not hgvs:
        results.append({
            "rsid": rsid,
            "hgvs": hgvs,
            "hgvs_exists_in_dbsnp": None,
            "dbsnp_hgvs_list": None
        })
        continue

    exists, hgvs_list = check_hgvs_in_dbsnp(rsid, hgvs)
    results.append({
        "rsid": rsid,
        "hgvs": hgvs,
        "hgvs_exists_in_dbsnp": exists,
        "dbsnp_hgvs_list": ",".join(hgvs_list)
    })

# 결과 DataFrame
outdf = pd.DataFrame(results)

# TSV로 저장
outdf.to_csv("/content/sample_data/litvar_validated_hgvs.tsv", sep="\t", index=False)

print(outdf.head())
```

```python
##litvar 최종코드
import ast
import json
import math
import logging
import re
import time
from collections import defaultdict
from urllib.parse import quote
from xml.etree import ElementTree

import pandas as pd
import requests
from tqdm.auto import tqdm


# ===============================
# 0. 공통 설정
# ===============================

logging.basicConfig(level=logging.INFO)
GNOMAD_API = "https://gnomad.broadinstitute.org/api"


# ===============================
# 1. LitVar gene 기반 수집
# ===============================

class LitVarFetcher:
    def __init__(self, gene_name: str):
        self.gene_name = gene_name
        self.session = requests.Session()

    # 1-1) gene → variant 리스트
    def fetch_gene_variants(self):
        base_url = "https://www.ncbi.nlm.nih.gov/research/litvar2-api/variant/search/gene/"
        url = base_url + self.gene_name

        try:
            response = self.session.get(url, timeout=20)
            response.raise_for_status()
        except Exception as e:
            logging.error(f"LitVar gene search API 오류: {e}")
            return []

        variants = []
        lines = response.text.strip().split("\n")
        for line in lines:
            line = line.strip()
            if not line:
                continue
            try:
                d = ast.literal_eval(line)
                variants.append(d)
            except Exception:
                continue

        return variants

    # 1-2) variant → PMID 리스트
    def fetch_pmids_for_variant(self, variant_id: str):
        base_url = "https://www.ncbi.nlm.nih.gov/research/litvar2-api/variant/get/"
        encoded = quote(variant_id)
        url = f"{base_url}{encoded}/publications"

        try:
            r = self.session.get(url, timeout=10)
            r.raise_for_status()
            j = r.json()
            pmids = j.get("pmids", [])
            if not pmids:
                logging.info(f"{variant_id}에는 PMID가 없습니다.")
            return pmids
        except Exception as e:
            logging.warning(f"PMID 조회 실패 ({encoded}): {e}")
            return []

    # 1-3) variant → hgvs 리스트
    def fetch_hgvs_for_variant(self, variant_id: str):
        base_url = "https://www.ncbi.nlm.nih.gov/research/litvar2-api/variant/get/"
        encoded = quote(variant_id)
        url = f"{base_url}{encoded}"

        try:
            r = self.session.get(url, timeout=10)
            r.raise_for_status()
            j = r.json()
            hgvs = j.get("hgvs", [])
            if not hgvs:
                logging.info(f"{variant_id}에는 hgvs가 없습니다.")
            return hgvs
        except Exception as e:
            logging.warning(f"hgvs 조회 실패 ({encoded}): {e}")
            return []

    # 1-4) LitVar → DataFrame(_id, rsid, gene, pmid, hgvs)
    def build_variant_dataframe(self):
        variants = self.fetch_gene_variants()
        if not variants:
            logging.error("변이 정보를 가져오지 못했습니다.")
            return pd.DataFrame()

        rows = []

        for v in tqdm(variants, desc="LitVar PMID/HGVS 수집 중"):
            vid = v.get("_id")
            rsid = v.get("rsid")
            gene_list = v.get("gene", [])

            if not vid:
                continue

            pmids = self.fetch_pmids_for_variant(vid)
            hgvs = self.fetch_hgvs_for_variant(vid)

            gene_str = ",".join(gene_list)
            hgvs_val = hgvs if hgvs else None  # 전체 리스트를 한 컬럼에 넣음

            if not pmids:
                rows.append({
                    "_id": vid,
                    "rsid": rsid,
                    "gene": gene_str,
                    "pmid": None,
                    "hgvs": hgvs_val
                })
            else:
                for pmid in pmids:
                    rows.append({
                        "_id": vid,
                        "rsid": rsid,
                        "gene": gene_str,
                        "pmid": int(pmid),
                        "hgvs": hgvs_val
                    })

            time.sleep(0.3)  # API 부담 완화

        df = pd.DataFrame(rows)
        return df


# ===============================
# 2. PMID 병합 및 개수 컬럼
# ===============================

def merge_pmids(df: pd.DataFrame) -> pd.DataFrame:
    df_merged = (
        df.groupby(['_id', 'rsid'], as_index=False)
          .agg({
              'pmid': lambda x: ",".join(sorted(set(x.dropna().astype(str)))),
              'gene': lambda x: ",".join(sorted(set(x.dropna().astype(str)))),
              'hgvs': lambda x: x.iloc[0]  # 동일 variant면 같은 hgvs라고 가정
          })
    )
    df_merged['pmids_count'] = df_merged['pmid'].apply(
        lambda s: len(s.split(',')) if isinstance(s, str) and s else 0
    )
    return df_merged


# ===============================
# 3. gnomAD / dbSNP / VEP 관련 유틸
# ===============================

def clean_value(x):
    if x is None:
        return ""
    if isinstance(x, float) and (math.isnan(x) or math.isinf(x)):
        return ""
    return str(x).strip()


def check_dbsnp(rsid):
    if not rsid.startswith("rs"):
        return False
    rs_num = rsid.replace("rs", "")
    if not rs_num.isdigit():
        return False

    url = f"https://api.ncbi.nlm.nih.gov/variation/v0/beta/refsnp/{rs_num}"
    try:
        r = requests.get(url, timeout=5)
        return r.status_code == 200
    except Exception:
        return False


def query_gnomad(rsid=None, hgvs=None, ref="GRCh38"):
    rsid = rsid if rsid else None
    hgvs = hgvs if hgvs else None

    if rsid is None and hgvs is None:
        return None

    query = """
    query Variant($rsid: String, $hgvs: String, $ref: ReferenceGenomeId!) {
      variant(rsid: $rsid, hgvs: $hgvs, referenceGenome: $ref) {
        variantId
        chrom
        pos
        ref
        alt
        genome { af ac an }
        exome { af ac an }
      }
    }
    """
    variables = {"rsid": rsid, "hgvs": hgvs, "ref": ref}

    try:
        r = requests.post(GNOMAD_API, json={"query": query, "variables": variables}, timeout=10)
        return r.json().get("data", {}).get("variant")
    except Exception:
        return None


def vep_translate(hgvs, gene):
    hgvs_full = f"{gene}:{hgvs}"  # gene:p.XxxY
    url = f"https://rest.ensembl.org/vep/human/hgvs/{hgvs_full}"
    try:
        r = requests.get(url, headers={"Content-Type": "application/json"}, timeout=10)
        if r.status_code != 200:
            return None
        return r.json()
    except Exception:
        return None


def process_variant(rsid, gene, hgvs):
    rsid = clean_value(rsid)
    gene = clean_value(gene)
    hgvs = clean_value(hgvs)

    dbsnp_exists = False
    gnomad_result = None
    mapped_hgvs = None

    # 3-1) rsID로 먼저 조회
    if rsid.startswith("rs"):
        dbsnp_exists = check_dbsnp(rsid)
        gnomad_result = query_gnomad(rsid=rsid)

    # 3-2) rsID로 안 되면 HGVS로 gnomAD 조회
    if not gnomad_result and hgvs:
        mapped_hgvs = hgvs
        gnomad_result = query_gnomad(hgvs=hgvs)

        # p.HGVS → c.HGVS 변환 시도
        if not gnomad_result and hgvs.startswith("p.") and gene:
            vep_output = vep_translate(hgvs, gene)
            if isinstance(vep_output, list):
                for entry in vep_output:
                    for t in entry.get("transcript_consequences", []):
                        hgvsc = t.get("hgvsc")
                        if hgvsc:
                            mapped_hgvs = hgvsc
                            gnomad_result = query_gnomad(hgvs=hgvsc)
                            if gnomad_result:
                                break

    return {
        "rsid": rsid,
        "gene": gene,
        "input_hgvs": hgvs,
        "mapped_hgvs": mapped_hgvs,
        "dbsnp_exists": dbsnp_exists,
        "gnomad_exists": gnomad_result is not None,
        "gnomad_variantId": gnomad_result["variantId"] if gnomad_result else None,
        "gnomad_af_genome": gnomad_result["genome"]["af"] if (gnomad_result and gnomad_result["genome"]) else None,
        "gnomad_af_exome": gnomad_result["exome"]["af"] if (gnomad_result and gnomad_result["exome"]) else None
    }


# ===============================
# 4. dbSNP HGVS 수집 및 매칭
# ===============================

def get_hgvs_from_dbsnp(rsid):
    """
    rsID에 대한 c.HGVS, p.HGVS 리스트 둘 다 반환
    """
    rs_num = rsid.replace("rs", "")
    url = f"https://api.ncbi.nlm.nih.gov/variation/v0/beta/refsnp/{rs_num}"
    r = requests.get(url, timeout=10)
    if r.status_code != 200:
        return [], []

    data = r.json()
    primary = data.get("primary_snapshot_data", {})

    hgvsc_set = set()
    hgvsp_set = set()

    for placement in primary.get("placements_with_allele", []):
        for allele in placement.get("alleles", []):
            for hg in allele.get("hgvs", []):
                if "coding" in hg:
                    hgvsc_set.add(hg["coding"])
                if "protein" in hg:
                    hgvsp_set.add(hg["protein"])

    return list(hgvsc_set), list(hgvsp_set)


def extract_c_hgvs(hgvs_list):
    if not hgvs_list:
        return ''
    only_c = [x.split(':', 1)[1] if ':' in x else x for x in hgvs_list]
    return ', '.join(only_c)


def extract_p_hgvs(hgvs_list):
    if not hgvs_list:
        return ''
    only_p = [x.split(':', 1)[1] if ':' in x else x for x in hgvs_list]
    return ', '.join(only_p)


aa_1to3 = {
    "A": "Ala", "R": "Arg", "N": "Asn", "D": "Asp", "C": "Cys",
    "Q": "Gln", "E": "Glu", "G": "Gly", "H": "His", "I": "Ile",
    "L": "Leu", "K": "Lys", "M": "Met", "F": "Phe", "P": "Pro",
    "S": "Ser", "T": "Thr", "W": "Trp", "Y": "Tyr", "V": "Val",
    "*": "Ter"
}


def one_to_three(hgvs):
    """
    p.R423C → p.Arg423Cys
    """
    if hgvs.startswith('p.') and len(hgvs) >= 6:
        m = re.match(r"p\.([A-Z])(\d+)([A-Z*])", hgvs)
        if m:
            ref1, pos, alt1 = m.groups()
            ref3 = aa_1to3.get(ref1, ref1)
            alt3 = aa_1to3.get(alt1, alt1)
            return f"p.{ref3}{pos}{alt3}"
    return hgvs


def check_dbsnp_match_row(row):
    hgvs_val = str(row['hgvs']).strip()
    if not hgvs_val or hgvs_val.lower() == 'nan':
        return '', ''

    hgvs_3letter = one_to_three(hgvs_val)

    hgvsc_list = str(row.get('only_dbsnp_hgvsc', '')).split(', ') if row.get('only_dbsnp_hgvsc') else []
    hgvsp_list = str(row.get('only_dbsnp_hgvsp', '')).split(', ') if row.get('only_dbsnp_hgvsp') else []

    if hgvs_3letter in hgvsc_list or hgvs_3letter in hgvsp_list:
        return 'o', hgvs_3letter
    else:
        return 'x', hgvs_3letter


# ===============================
# 5. PMID 유효성 체크(PubMed)
# ===============================

def check_pmid_exists(pmid):
    url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi"
    params = {
        "db": "pubmed",
        "id": pmid,
        "rettype": "abstract",
        "retmode": "xml"
    }
    try:
        r = requests.get(url, params=params, timeout=10)
        if r.status_code != 200:
            return False
        root = ElementTree.fromstring(r.content)
        article = root.find(".//PubmedArticle")
        return (article is not None)
    except Exception:
        return False


# ===============================
# 6. 전체 파이프라인 실행
# ===============================

def run_pipeline(gene: str, out_prefix: str = "litvar2"):
    # 6-1) LitVar 수집 → pmid_hgvs.txt
    fetcher = LitVarFetcher(gene)
    df_raw = fetcher.build_variant_dataframe()
    df_raw.to_csv(f"{out_prefix}_{gene}_pmid_hgvs.txt", sep="\t", index=False)

    # 6-2) PMID 병합
    df_merged = merge_pmids(df_raw)
    df_merged.to_csv(f"{out_prefix}_{gene}_pmid_merge.txt", sep="\t", index=False)

    # 6-3) gnomAD / dbSNP 존재 여부 등
    results = []
    for _, row in tqdm(df_merged.iterrows(), total=len(df_merged), desc="gnomAD/dbSNP 검증 중"):
        res = process_variant(
            rsid=row.get("rsid", ""),
            gene=row.get("gene", ""),
            hgvs=row.get("hgvs", "")
        )
        results.append(res)

    outdf = pd.DataFrame(results)

    # rsID 없는 행 제거
    outdf = outdf[outdf["rsid"].str.strip() != ""].copy()

    # 6-4) dbSNP HGVS 리스트 가져와 litvar_df와 병합
    litvar_df = df_merged.copy()
    litvar_df = litvar_df[litvar_df["rsid"].notna()].copy()

    litvar_df['dbsnp_hgvsc'] = None
    litvar_df['dbsnp_hgvsp'] = None

    for rsid in tqdm(litvar_df['rsid'].unique(), desc="dbSNP HGVS 수집 중"):
        hgvsc_list, hgvsp_list = get_hgvs_from_dbsnp(rsid)
        idx = litvar_df[litvar_df['rsid'] == rsid].index
        for i in idx:
            litvar_df.at[i, 'dbsnp_hgvsc'] = hgvsc_list
            litvar_df.at[i, 'dbsnp_hgvsp'] = hgvsp_list

    # dbSNP HGVS에서 접두사 제거
    litvar_df['only_dbsnp_hgvsc'] = litvar_df['dbsnp_hgvsc'].apply(extract_c_hgvs)
    litvar_df['only_dbsnp_hgvsp'] = litvar_df['dbsnp_hgvsp'].apply(extract_p_hgvs)

    # LitVar HGVS와 dbSNP HGVS 매칭
    litvar_df[['dbsnp_match', 'litvar_hgvs']] = litvar_df.apply(
        check_dbsnp_match_row, axis=1, result_type='expand'
    )

    # 6-5) PMID 유효성 체크(옵션)
    # 예: pmid 컬럼에서 첫 번째 pmid만 골라 확인
    litvar_df['first_pmid'] = litvar_df['pmid'].apply(
        lambda s: s.split(',')[0] if isinstance(s, str) and s else None
    )
    litvar_df['pmid_exists'] = litvar_df['first_pmid'].apply(
        lambda x: check_pmid_exists(x) if x else None
    )

    # 6-6) gnomAD 결과와 병합
    final_df = litvar_df.merge(outdf, on=['rsid', 'gene'], how='left', suffixes=('', '_gnomad'))

    out_file = f"{out_prefix}_{gene}_match_res.tsv"
    final_df.to_csv(out_file, sep="\t", index=False)
    print(f"\n=== 완료: {out_file} 저장됨 ===")

    return final_df


# 실제 실행 예시
if __name__ == "__main__":
    gene = "EEF1A2"
    final_df = run_pipeline(gene)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn3, venn2

# 데이터 로드
df=pd.read_excel('/content/sample_data/CLINVAR_res.xlsx')

def make_set(col):
    return set(df.loc[df[col] == 'O', 'hgvsc'])

set_web = make_set('clinvar web')
set_api = make_set('clinvar api')
set_yoon = make_set('윤교수님')
set_lit = make_set('litvar')

# Figure 1: ClinVar Web vs API vs 윤교수님
plt.figure(figsize=(6,6))
venn3(
    [set_web, set_api, set_yoon],
    set_labels=('ClinVar Web', 'ClinVar API', '윤교수님')
)
plt.title("Overlap of variants (HGVSc)\nClinVar Web vs API vs 윤교수님")
plt.show()

# Figure 2: ClinVar (Web ∪ API) vs LitVar
clinvar_union = set_web.union(set_api)

plt.figure(figsize=(6,6))
venn2(
    [clinvar_union, set_lit],
    set_labels=('ClinVar (Web ∪ API)', 'Litvar')
)
plt.title("Overlap of variants (HGVSc)\nClinVar vs Litvar")
plt.show()
```
